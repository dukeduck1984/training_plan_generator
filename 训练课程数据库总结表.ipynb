```python
from openpyxl import Workbook
from openpyxl import load_workbook
from openpyxl.compat import range
from openpyxl.styles import Font, Color, Border, Side, Alignment, Protection
from openpyxl.utils import get_column_letter
import numpy as np
import pandas as pd
from pandas import DataFrame
import random
import arrow

wko_db_path = "D:/客户训练计划/AutoGenerated/WKO_DB/workout_db.xlsx"

sports = ["swim", "bike", "run", "brick"]
races = ["sprint", "olympic", "half", "full"]
stages = ["prep", "base", "build", "peak"]
types = ["ae", "ss", "me", "fo", "bk", "re"]

for race in races:
    print("比赛距离：" + race.title())
    for stage in stages:
        print("\t周期阶段：" + stage.title())
        for sport in sports:
            print("\t\t运动：" + sport.title())
            df = pd.read_excel(wko_db_path, sheetname = sport)
            selected_stage = df[df['stage']==stage]
            selected_sport = selected_stage[selected_stage['race']==race]
            for wtype in types:
                selected_type = selected_sport[selected_sport['type']==wtype]
                type_count = selected_type['workout'].count()
                if type_count != 0:
                    print("\t\t\t训练类型：" + wtype.upper(), end="")
                    print("      数量：" + str(type_count), end="")
                    get_duration = selected_type['duration']
                    min_d = min(get_duration)
                    max_d = max(get_duration)
                    print("      时长：" + str(min_d) + " - " + str(max_d))
```
